The value counts in some of the feaures are different in the training and testing data so we need to add these features to the training data.

```python
#before that we must drop the the features that have more than half missing features
#these are the columns where most of the data are missing
drop_columns = df_test.columns[df_test.isnull().sum() > len(df_test)/2].tolist()
drop_columns
```

```python
df_test.drop(drop_columns, axis = 1,inplace = True)
#updating the categorical_missing_features variable
categorical_missing_columns = list(set(categorical_missing_columns) - set(drop_columns))

df_test.isnull().sum()
```

For now we are going to fill the missing values in the test data's categorical features using the mode of the respective column and we will fill the missing values in the continuous columns with the mean of the respective columns.

```python
#filling the missing categorical variables with the mode
for column in categorical_missing_columns:
    df_test[column] = df_test[column].fillna(df_test[column].mode()[0])
```

```python
for column in numeric_missing_columns:
    df_test[column] = df_test[column].fillna(df_test[column].mode()[0])
```
